# Cell 5  Point the task config to your dataset

```python
from pathlib import Path
import yaml

CONFIG_PATH = Path("/kaggle/working/dps/configs/super_resolution_config.yaml")
DATA_ROOT   = Path("/kaggle/input/ffhq256-images-only/ffhq256")  # your exact path

with open(CONFIG_PATH, "r") as f:
    cfg = yaml.load(f, Loader=yaml.FullLoader)   # safe_load fails on !!python/tuple

cfg["data"]["root"] = str(DATA_ROOT)

with open(CONFIG_PATH, "w") as f:
    yaml.dump(cfg, f, sort_keys=False, Dumper=yaml.FullDumper)

print("Updated data.root ->", cfg["data"]["root"])
```

# Cell 6  (Optional) Test on a single file first

```python
from pathlib import Path
import shutil

sample_dir = Path("/kaggle/working/dps/data/samples")
sample_dir.mkdir(parents=True, exist_ok=True)

# pick the first image
first_img = sorted(Path(cfg["data"]["root"]).rglob("*.png"))[:1]
if not first_img:
    first_img = sorted(Path(cfg["data"]["root"]).rglob("*.jpg"))[:1]
assert first_img, "No images found under data.root"

dst = sample_dir / first_img[0].name
shutil.copy(first_img[0], dst)
print("Sample ->", dst)

# temporarily point config to the samples dir (one-image smoke test)
import yaml
with open(CONFIG_PATH, "r") as f:
    cfg2 = yaml.load(f, Loader=yaml.FullLoader)
cfg2["data"]["root"] = str(sample_dir)
with open(CONFIG_PATH, "w") as f:
    yaml.dump(cfg2, f, sort_keys=False, Dumper=yaml.FullDumper)
print("Temporarily set data.root to:", cfg2["data"]["root"])
```
